```
import pandas
import os
#VAMOS A EXTRAER DATOS DE UNA PAGINA WEB https://pokemondb.net/pokedex/all  UTILIZANDO  import requests
import requests
```

```
#EXTRAEMOS EL CODIGO
url = "https://pokemondb.net/pokedex/all"
html_text = requests.get(url).content
```

```
 #CREAMOS UN DATAFREAM
contenido = pandas.read_html(html_text)
```

```
# extraemos los datos
df_pokedex = contenido[0]
```

```
#VAMOS A SUMAR LA COLUMNAS DE ATAQUE Y ATAQUE ESPECIAL CON EL SIGUIENTE COMANDO
df_pokedex["Attack"] + df_pokedex["Sp. Atk"]
```

```
#PARA ASIGNARLOS A UNA NUEVA COLUMNA (TOTAL_ataque)
df_pokedex["Total_ataque"] =  df_pokedex["Attack"] + df_pokedex["Sp. Atk"]
```

```
#PARA SUMAR TODAS LAS COLUMNAS DE LA TABLA
df_pokedex.get(["HP", "Attack","Defense", "Sp. Atk", "Sp. Def", "Speed"]).sum(axis = 1)
```

```
#PARA ASIGNARLOS A UNA NUEVA COLUMNA 
df_pokedex["Total_ Estadistica"] = df_pokedex.get(["HP", "Attack","Defense", "Sp. Atk", "Sp. Def", "Speed"]).sum(axis = 1)
```
